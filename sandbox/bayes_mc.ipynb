```python
import pandas as pd
import numpy as np
from scipy.stats import binom
```

```python

tot_samples = 100000

df = pd.DataFrame(index = range(tot_samples), columns = ['prob', 'prob_game', 'rand1', 'game', 'prob_win', 'rand2', 'win'])

df['rand1'] = np.random.random(len(df['rand1']))
df['rand2'] = np.random.random(len(df['rand2']))

# PRIOR PORBA = 3/8
# Create simplified MCMC 

df.loc[df.index <10000, 'prob'] = 0.375
df.loc[df.index <10000,'prob_game'] = binom.pmf(3, 8, 0.375)
df.loc[df.index >= 10000, 'prob'] = 0.6
df.loc[df.index >= 10000,'prob_game'] = binom.pmf(3, 8, 0.6)

df['game'] = df['rand1'] < df['prob_game']


df['prob_win'] = df['prob'].pow(3)


df['win'] = df['rand2'] < df['prob_win']
df.head()
```
