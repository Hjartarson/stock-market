```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
import os
import matplotlib as mpl

import os
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ska_years.csv', skiprows=1, delimiter=';')
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by='Date')
df['Closing price'] = df['Closing price'].str.replace(',','.').astype('float')
df = df.set_index('Date')
df#['Handle'] = df['Price'].mul(df['Volume'])
```

```python
#df['Closing price'] = df['Closing price'].str.replace(',','.').astype('float')
f, ax = plt.subplots(figsize = (14,10))
df['Closing price'].plot(ax=ax)
df['Closing price'].rolling(10).mean().plot(ax=ax, c='red')
df['Closing price'].rolling(20).mean().plot(ax=ax)
df['Trades'].plot(ax=ax, secondary_y = True)

```

```python
df['pct_change'] = df['Closing price'].pct_change()
df.head()
```

```python
f, ax = plt.subplots(figsize = (14,10))

df['pct_change'].rolling(10).mean().plot(ax=ax, ls='--')
df['pct_change'].rolling(10).median().plot(ax=ax)
df['Closing price'].plot(ax=ax, secondary_y = True)
ax.axhline(0, c='r', ls='--')
```
